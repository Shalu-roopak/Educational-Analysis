<h3>Analysing highest literacy rate of 2011 to rural vs urban</h3>

```python
data.head(2)
```

```python

x_data = data['Country/ States/ Union Territories Name']
y_data = data['Literacy Rate (Persons) - Rural - 2011']
plt.scatter(x_data,y_data,color='b',label='Rural')


y=data['Literacy Rate (Persons) - Urban - 2011']
plt.scatter(x_data,y,color='r',label='Urban')
plt.xticks(rotation=90)
plt.xlabel('Country/State/UT')
plt.ylabel('Literacy Rate (Persons) - 2011')
plt.title('Literacy Rate Comparison: Rural vs Urban (2011)')

plt.legend()

plt.show()
```

<p style="font-size:18px">from the initial observance of the data ,the first index or the row numbered <b>0</b> represents the details of overall country. And easily understands that the  <b style="color:green">total literacy rate of 2011 of <i>rural</i> is 67.8 and<i> urban</i> is 84.1</b> .Thus knows that urban has more literacy rate .</p>

## Analysing highest literacy rate of state

```python
data.head(2)
```

```python
data_refined=data.drop(0)
data_refined
```

```python
data_refined.columns
```

```python
plt.figure(figsize=(15,10))
plt.title("Analysing highest literacy rate of state", fontweight="bold")
x_data = data_refined['Country/ States/ Union Territories Name']
y_data = data_refined['Literacy Rate (Persons) - Total - 2011']
cmap = plt.get_cmap('plasma')
norm = plt.Normalize(vmin=min(y_data), vmax=max(y_data))
bar_colors = cmap(norm(y_data))
plt.bar(x_data,y_data,width=0.6,color=bar_colors)
plt.xticks(rotation=90)
plt.show()
```

<p style="font-size:18px;"> From the above analysis found out that <b style="color:red;"> kerala</b> has the  highest literacy rate 2011,Mizoram and Lakshadweep has came to the second literacy rate in 2011</p> 

## Comparison  of  rural and urban areas in each state for the years 2011 and 2001


```python
data_refined.head(2)
```

```python
plt.figure(figsize=(15,10))
plt.title("Comparison of rural literacy of 2001 vs 2011", fontweight="bold")
plt.xlabel("states")
plt.ylabel("literacy rate(%)")

xdata = data_refined['Country/ States/ Union Territories Name']
ydata=data_refined['Literacy Rate (Persons) - Rural - 2001']
plt.plot(xdata,ydata,label="2001",marker='o',mfc='y',mec='r')

y2data=data_refined['Literacy Rate (Persons) - Rural - 2011']
plt.plot(xdata,y2data,label="2011",color='g',marker='*',mfc='r',mec='r',ls='--')
plt.xticks(rotation=90)
plt.grid()
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title("comparison of urban literacy of 2001 vs 2011", fontweight="bold")
plt.xlabel("states")
plt.ylabel("literacy rate(%)")

xdata = data_refined['Country/ States/ Union Territories Name']
ydata=data_refined['Literacy Rate (Persons) - Urban - 2001']
plt.plot(xdata,ydata,label="2001",marker='o',mfc='y',mec='r')

y2data=data_refined['Literacy Rate (Persons) - Urban - 2011']
plt.plot(xdata,y2data,label="2011",color='g',marker='*',mfc='r',mec='r',ls='--')
plt.xticks(rotation=90)
plt.grid()
plt.legend()

plt.show()
```

 <p style="font-size:18px; line-height:1.6; margin:10px 0;">
From the analysis, it was found that states and territories with consistently high or low literacy rates can be identified. <b style="color:red;">Kerala</b> recorded the highest literacy rate in rural areas in 2011, while <b style="color:orange;">Mizoram </b>had the highest literacy rate in urban areas in 2011. A comparative diagram highlights that Kerala's total literacy rate in rural areas was 93%, whereas the urban literacy rate in Mizoram was 97.2%. This comparison demonstrates that urban areas generally have higher literacy rates than rural areas.
</p>
